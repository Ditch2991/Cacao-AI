```python
import requests
from IPython.display import Image, display

# URL de l'endpoint de prédiction
url = "https://amouss3e-cocoa-diseases-api.hf.space/predict"

# Chemin de l'image locale
image_path = "Images/image1.jpg"

# En-têtes avec le token d'authentification Hugging Face
headers = {
    "Authorization": "Bearer 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"  
}

# Lecture et envoi de l'image
with open(image_path, 'rb') as img:
    files = {'file': (image_path, img, 'image/jpeg')}  
    response = requests.post(url, headers=headers, files=files)

# Résultat de la prédiction
if response.status_code == 200:
    with open("result.png", "wb") as f:
        f.write(response.content)
    print("✅ Image prédite enregistrée sous result.png")


# Affiche l'image de prédiction
display(Image(filename="result.png"))
```
